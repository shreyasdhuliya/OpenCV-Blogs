# kernels

```python
#importing libs
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel = []
kernel_name = []
kernel.append(np.ones((3,3),np.uint8))
kernel_name.append("np.ones(3,3)")
kernel.append(np.ones((5,5),np.uint8))
kernel_name.append("np.ones(5,5)")
kernel.append(cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)))
kernel_name.append("MORPH_RECT(5,5)")
kernel.append(cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)))
kernel_name.append("MORPH_ELLIPSE(5,5)")                   
kernel.append(cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)))
kernel_name.append("MORPH_CROSS(5,5)") 
kernel.append(cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3)))
kernel_name.append("MORPH_CROSS(3,3)")
```

```python
for i in range(len(kernel)):
    print(kernel[i])
```

# Tansformations

## 1. Erosion

```python
img = cv2.imread("assets/A.png")
plt.imshow(img)
```

```python

fig = plt.figure(figsize = (36, 26)) 
count = 1
for i in range(1,5,1):
    for j in range(1,len(kernel)+1):
        img_c = img.copy()
        erode = cv2.erode(img_c,kernel[j-1],iterations = i)
        ax = fig.add_subplot(4, 6, count)
        count = count + 1 
        ax.imshow(erode,cmap = 'gray')
        ax.set_title(str(kernel_name[j-1] + " iter = " +str(i)),fontsize=20)
```

## 2. Dialation

```python
img = cv2.imread("assets/A_thin.png")
plt.imshow(img)
```

```python
fig = plt.figure(figsize = (36, 26)) 
count = 1
for i in range(1,5,1):
    for j in range(1,len(kernel)+1):
        img_c = img.copy()
        erode = cv2.dilate(img_c,kernel[j-1],iterations = i)
        ax = fig.add_subplot(4, 6, count)
        count = count + 1 
        ax.imshow(erode,cmap = 'gray')
        ax.set_title(str(kernel_name[j-1] + " iter = " +str(i)),fontsize=20)
```

## 3. Open

```python
img = cv2.imread("assets/A_open.png")
fig = plt.figure(figsize = (10, 10)) 
ax = fig.add_subplot(1, 2, 1)
ax.imshow(img, cmap = "gray")
ax = fig.add_subplot(1, 2, 2)
img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel[0]) 
ax.imshow(img, cmap = "gray")
```

## 4. Close 

```python
img = cv2.imread("assets/A_close.png")
fig = plt.figure(figsize = (10, 10)) 
ax = fig.add_subplot(1, 2, 1)
ax.imshow(img, cmap = "gray")
ax = fig.add_subplot(1, 2, 2)
img = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel[0]) 
ax.imshow(img, cmap = "gray")
```
